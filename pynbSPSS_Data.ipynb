```python
import pandas as pd
import pyreadstat
import matplotlib.pyplot as plt
import seaborn.apionly as sns 
```

```python
df = pd.read_spss("/home/auto/Codes/Dengue/total.sav")
df, meta = pyreadstat.read_sav('/home/auto/Codes/Dengue/total.sav')
```

```python
df = df.drop("train",1)
```

```python
df
```

```python
dengue_avg = df.groupby('moy').dcase.mean()
dengue_avg
```

```python
plt.figure(figsize = (10,10))
ax = plt.gca()
dengue_avg.plot(kind="barh")
plt.grid(axis = 'x', color ='white', linestyle='-')
plt.xlabel("Dengue Case in Average for month")
ax.tick_params(axis='both', which='both',length=0)
sns.despine()
plt.savefig('DCaseAvg.png')
```

```python
dengue_avg_dhk = df[df.site==1.0]
dengue_avg_dhk = dengue_avg_dhk.groupby('moy').dcase.mean()
type(dengue_avg_dhk)
```

```python
plt.figure(figsize = (10,10))
ax = plt.gca()
dengue_avg_dhk.plot(kind="barh")
plt.grid(axis = 'x', color ='white', linestyle='-')
plt.xlabel("Dengue Case in Average for month for Dhaka")
ax.tick_params(axis='both', which='both',length=0)
sns.despine()
plt.savefig("DCaseAvgDhaka.png")
```

```python
dengue_avg_ctg = df[df.site==2.0]
dengue_avg_ctg = dengue_avg_ctg.groupby('moy').dcase.mean()
```

```python
plt.figure(figsize = (10,10))
ax = plt.gca()
dengue_avg_ctg.plot(kind="barh")
plt.grid(axis = 'x', color ='white', linestyle='-')
plt.xlabel("Dengue Case in Average for month for Chittagong")
ax.tick_params(axis='both', which='both',length=0)
sns.despine()
plt.savefig("DCaseAvgChittagong.png")
```

```python
dengue_avg_dhk.plot(style='.-',label="Dhaka")
dengue_avg_ctg.plot(style='.-',label="Chittagong")
plt.savefig("DCaseAvgCompareDhakaToChittagong.png")
```

```python
dengue_ctg = df[df.site==2.0]
dengue_ctg = pd.Series(dengue_ctg['dcase'].values,dengue_ctg['mid'].values)
dengue_ctg.plot(style='.-',figsize=(20,10),color='b')

dengue_dhk = df[df.site==1.0]
dengue_dhk_dcase = pd.Series(dengue_dhk['dcase'].values,dengue_dhk['mid'].values)
dengue_dhk_dcase.plot(style='.-',figsize=(20,10),color='k')
```

```python
dengue_dhk = df[df.site==1.0]
dengue_dhk_temp = pd.Series(dengue_dhk['mtem'].values,dengue_dhk['mid'].values)
x = dengue_dhk_dcase.plot(style='.-',figsize=(20,10),color='b')
x2 = x.twinx()
dengue_dhk_temp.plot(style='.-',figsize=(20,10),color='k', x=x2)
plt.savefig("DCaseDHKTemp.png")
```

```python
dengue_dhk = df[df.site==1.0]
dengue_dhk_temp = pd.Series(dengue_dhk['mhumi'].values,dengue_dhk['mid'].values)
x = dengue_dhk_dcase.plot(style='.-',figsize=(20,10),color='k',label='dengue case',marker='.')
x2 = x.twinx()
dengue_dhk_temp.plot(style='.-',figsize=(20,10),color='g', x=x2)
plt.savefig("DCaseDHKTemp.png")
```
